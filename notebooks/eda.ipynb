```python
categorical_cols = [
    "CurrencyCode",
    "CountryCode",
    "ProviderId",
    "ProductId",
    "ProductCategory",
    "ChannelId",
    "PricingStrategy",
    "FraudResult"
]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in categorical_cols:
    print("\n", col)
    print(df[col].value_counts())

    plt.figure(figsize=(8,4))
    sns.countplot(x=df[col])
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
print(numeric_cols)
```

```python
corr = df[numeric_cols].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
for col in ["Amount", "Value"]:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```

```python
for col in ["Amount", "Value"]:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

## Key Insights

1. Fraud cases are highly imbalanced compared to non-fraud transactions.
2. Transaction amounts are highly skewed with extreme outliers.
3. ProductCategory shows strong concentration in a few categories.
4. Some numerical features (Amount, Value) are highly correlated.
5. Certain channels/providers dominate transaction volume.
